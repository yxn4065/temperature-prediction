```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 加载数据  
df = pd.read_csv('./Lingcang.csv')

# 数据的基本统计信息  
print(df.describe())

# 设置plt字体
# 设置全局字体为SimHei（黑体），你需要确保你的系统中安装了这个字体  
plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体为黑体  
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像时负号'-'显示为方块的问题 
```

```python
# 绘制最高气温的时间序列图  
plt.figure(figsize=(12, 6))
plt.plot(df['year'] + (df['month'] - 1) / 12.0 + df['day'] / 365.0, df['最高气温'], marker='o')
plt.title('最高气温时间序列')
plt.xlabel('时间（年）')
plt.ylabel('最高气温（℃）')
plt.grid(True)
plt.show()
```

```python
# 绘制最低气温和最高气温的箱线图，按年份分组  
plt.figure(figsize=(12, 6))
sns.boxplot(x='year', y='最低气温', data=df, palette="Set2")
plt.title('每年最低气温箱线图')
plt.xlabel('年份')
plt.ylabel('最低气温（℃）')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='year', y='最高气温', data=df, palette="Set2")
plt.title('每年最高气温箱线图')
plt.xlabel('年份')
plt.ylabel('最高气温（℃）')
plt.show()
```

```python
# 绘制气温的月平均变化图（假设数据是连续的每日数据）  
monthly_avg = df.groupby(['year', 'month']).mean().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(x='month', y='平均气温', hue='year', data=monthly_avg, palette="tab10", markers=True)
plt.title('月平均气温变化')
plt.xlabel('月份')
plt.ylabel('平均气温（℃）')
plt.show()
```

```python
# 气温分布直方图：
# 展示最高气温、最低气温或平均气温的分布情况。

# 绘制最低气温的直方图  
plt.figure(figsize=(10, 6))
plt.hist(df['最低气温'], bins=20, edgecolor='black', alpha=0.7)
plt.title('最低气温分布直方图')
plt.xlabel('最低气温 (℃)')
plt.ylabel('频数')
plt.grid(True)
plt.show()
```

```python
# 气温变化的散点图矩阵：
# 展示不同气温指标之间的关系。
# python
# 复制代码
# 绘制气温的散点图矩阵  
sns.pairplot(df[['最低气温', '最高气温', '平均气温']], kind="scatter", hue="平均气温",
             plot_kws=dict(s=50, edgecolor="white", linewidth=2.5))
plt.show()
```

```python
# 小提琴图：
# 展示不同年份或月份的气温分布情况。
# python
# 绘制每年最高气温的小提琴图  
plt.figure(figsize=(12, 6))
sns.violinplot(x='year', y='最高气温', data=df, palette="Set2")
plt.title('每年最高气温小提琴图')
plt.xlabel('年份')
plt.ylabel('最高气温 (℃)')
plt.show()
```
